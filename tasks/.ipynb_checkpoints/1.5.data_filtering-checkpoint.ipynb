```python
#Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?
students_performance.groupby('lunch').aggregate({'math score' : ['mean', 'var'],'reading score' : ['mean', 'var'], 'writing score' : ['mean', 'var']}) #сгруппировать по полу и посчитать среднее
```
