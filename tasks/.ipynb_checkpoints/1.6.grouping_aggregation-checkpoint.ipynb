```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv('../resources/StudentsPerformance.csv')
```

```python
students_performance = students_performance.rename(columns={
                                        'parental level of education' : 'parental_level_of_education', 
                                        'test preparation course' : 'test_preparation_course',
                                        'math score' : 'math_score',
                                        'reading score' : 'reading_score',
                                        'writing score' : 'writing_score'})
```

```python
students_performance.head()
```

```python
students_performance.groupby('gender', as_index=False)\
                            .mean(numeric_only=True)\
                            .rename(columns={'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score', 'writing_score' : 'mean_writing_score'})
```

```python
mean_scores = students_performance.groupby(['gender', 'race/ethnicity'])\
                            .mean(numeric_only=True)\
                            .rename(columns={'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score', 'writing_score' : 'mean_writing_score'})
```

```python
mean_scores #с мультииндексами
```

```python
mean_scores.unstack()
```

```python
mean_scores.index
```

```python
mean_scores.loc[[('female','group A'), ('female','group B')]]
```

```python
# вывести топ 5 женщин и мужчин в математике
students_performance.sort_values(['gender', 'math_score'], ascending=False).groupby('gender').head(5) 
```

```python
# создание новых колонок
students_performance['total_score'] = students_performance.math_score + students_performance.reading_score + students_performance.writing_score
```

```python
students_performance.head()
```

```python
students_performance = students_performance.assign(total_score_log=np.log(students_performance.total_score))
```

```python
students_performance.head()
```

```python
# удаление
students_performance = students_performance.drop('total_score', axis=1)
```

```python
# Аспирант Ростислав изучает метаболом водорослей и получил такую табличку.
# В ней он записал вид каждой водоросли, её род (группа, объединяющая близкие виды), 
# группа (ещё одно объединение водорослей в крупные фракции) и концентрации анализируемых веществ.

# Помогите Ростиславу найти среднюю концентрацию каждого из веществ в каждом из родов (колонка genus)! 
# Для этого проведите группировку датафрэйма, сохранённого в переменной concentrations, и примените метод, 
# сохранив результат в переменной mean_concentrations.
```

```python
concentrations = pd.read_csv("../resources/algae.csv")
```

```python
concentrations
```

```python
mean_concentrations = concentrations.groupby('genus').mean(numeric_only=True)
mean_concentrations
```

```python
# Пользуясь предыдущими данными, укажите чему равны минимальная, 
# средняя и максимальная концентрации аланина (alanin) среди видов рода Fucus.
```

```python
res = concentrations.loc[concentrations.genus=='Fucus'][['alanin']].describe().loc[['min','mean','max']]
res['alanin'].apply(lambda x: round(x,2))
```

```python
# Дисперсия концентрации цитрата в бурых, зелёных и красных водорослях
concentrations_describe = concentrations.groupby('group')['citrate'].describe()
concentrations_describe
```

```python
concentrations_describe.iloc[:,[2]].apply(lambda x: x**2)
```

```python
#Число видов бурых, зелёных и красных водорослей
concentrations.groupby('group').size()
```

```python
#Размах (максимум - минимум) сахарозы в бурых, зелёных и красных водорослях
res = concentrations.groupby('group')['sucrose'].describe()
res
```

```python
res['range']= res['max']-res['min']
```

```python
res[['range']]
```

```python
concentrations.groupby('group') \
    .agg({'citrate': 'var', 'sucrose': lambda x: x.max() - x.min(), 'glucose': 'size'}) \
    .round(2)
```
